# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])

fig, axs = plt.subplots(1, 2, figsize=(11, 4), sharey=True)

for ax, ts, label in zip(axs,
                          [time_series_0, time_series_1],
                          ["Week 1", "Week 2"]):
    ax.plot(range(7), ts, marker="o", color="steelblue", linewidth=2)
    ax.set_xticks(range(7))
    ax.set_xticklabels(days, fontsize=9)
    ax.set_title(label, fontweight="bold", fontsize=11)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

axs[0].set_ylabel("Number of visitors", fontsize=10)
fig.suptitle("Museum Visitors by Day — Separate Panels",
             fontweight="bold", fontsize=12, x=0.08, ha="left")
plt.subplots_adjust(wspace=0.08)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(range(7), time_series_0, marker="o", color="steelblue",  linewidth=2, label="Week 1")
ax.plot(range(7), time_series_1, marker="o", color="darkorange", linewidth=2, label="Week 2")

ax.set_xticks(range(7))
ax.set_xticklabels(days, fontsize=9)
ax.set_ylabel("Number of visitors", fontsize=10)
ax.set_title("Museum Visitors by Day — Overlaid",
             fontweight="bold", loc="left", fontsize=12)
ax.legend(frameon=False)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

**Advantages and disadvantages of each presentation:**

**Separate panels (side-by-side plots):**

- *Advantage:* Both panels share a y-axis (with `sharey=True`), so visitor counts are directly comparable in scale even across panels.
- *Disadvantage:* Direct day-to-day comparison between the two weeks is harder — the eye must travel between panels and hold one pattern in memory while viewing the other. Small differences are easy to miss.
- *Disadvantage:* Uses roughly twice the horizontal space for the same information.

**Overlaid lines (single plot):**

- *Advantage:* Wherever the lines cross or diverge, differences jump out. This is the right choice when the main question is "how did Week 2 differ from Week 1?"
- *Advantage:* More compact; both series share one axis so the scale is automatically consistent.
- *Disadvantage:* If the two lines are close together or cross repeatedly, the chart can become cluttered and harder to read, especially with more than two series.
- *Disadvantage:* The within-week shape of each individual series can be harder to appreciate because the reader’s attention is split between the two lines.

**Bottom line:** Use separate panels to tell a story *about each individual week*; use the overlaid view to tell a story *about the difference between weeks*.
